### Q1. Explain GET and POST methods.

#### Ans.GET Method. The GET method is used to request the server to GET the data or information. It should not be used with the requesting sensitive data, and there are certain limits to the requesting data. 
.
#### POST method. The POST method is used to send the data to the server.


## Q2. Why is request used in Flask?

Ans.The Request, in Flask, is an object that contains all the data sent from the Client to Server. This data can be recovered using the GET/POST Methods.
