### Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?

```python
Ans.Bokeh provides many options for customizing the appearance of a plot, including the axes, title, and legend. 
```

```python
from bokeh.plotting import figure, output_file, show

# Create a new plot
p = figure(title="Customizing Axes")

# Set x-axis properties
p.xaxis.axis_label = "X-axis label"
p.xaxis.major_label_orientation = "vertical"

# Set y-axis properties
p.yaxis.axis_label = "Y-axis label"
p.yaxis.axis_line_width = 2
p.yaxis.axis_line_color = "red"

# Add a line glyph to the plot
p.line([1, 2, 3, 4, 5], [6, 7, 2, 4, 5], line_width=2)

# Specify where the output will be displayed
output_file("customizing_axes.html")

# Show the plot
show(p)
```
